 Graphiques avec matplotlib

Matplotlib est une bibliothèque du langage de programmation Python destinée à tracer et visualiser des données sous formes de graphiques. Elle peut être combinée avec les bibliothèques python de calcul scientifique NumPy et SciPy. 
Pour utiliser les fonctionnalités offertes par matplotlib, il est nécessaire de charger certains modules. Le plus commun est le sous-module pyplot, auquel on attribue fréquemment l’alias plt :
* Ressources: (https://matplotlib.org/)

# Graphiques de Base : Pyplot

```python
# les imports habituels
import numpy as np
import matplotlib.pyplot as plt
```

### Graphiques simples

```python
# si je ne donne qu'une seule liste à plot
# alors ce sont les Y
plt.plot([10, 20, 25, 28])
# on peut aussi facilement ajouter une légende
# ici sur l'axe des y
plt.ylabel('some numbers')

plt.show()
```

Si on donne seulement *une* liste de valeurs, elles sont considérées comme les *Y*, les *X* étant les entiers en nombre suffisant et en commençant à 0.

```python
X = np.linspace(0, 2, 10)
y = X**2
plt.plot(X, y)
plt.show()
```

Un des graphiques que l’on rencontre très fréquemment est le nuage de points. Pour en réaliser un, on peut faire appel à la fonction scatter(), à qui l’on indique les cordonnées (x,y) des points ainsi que quelques paramètres optionnels de forme ou d’esthétisme

```python
plt.scatter(X, y)
plt.show()
```

### Styles Graphiques
Il existe beaucoup de styles a ajouter aux graphiques. Voici les plus importants a retenir :
- **c** : couleur de la ligne
- **lw** : epaisseur de la ligne (pour les graphiques plot)
- **ls** : style de la ligne (pour les graphiques plot)
- **size** : taille du point (pour les graphiques scatter)
- **marker** : style de points (pour les graphiques scatter)
- **alpha** : transparence du graphique



```python
plt.plot(X, y, c='red', lw=4, ls='--', alpha=0.9)
plt.show()
```

## Etapes classique pour la création d'une figure
Pour créer des figures proprements, on doit suivre le cycle de vie suivant :
1. **plt.figure**(figsize())
2. **plt.plot()**
3. Extras (titre, axes, legendes)
4. **plt.show()**

```python
X = np.linspace(0, 2, 10)

plt.figure(figsize=(5,10)) # Création d'une figure
plt.plot(X, X**2, label='quadratique') # premiere courbe
plt.plot(X, X**3, label='cubique') # deuxieme courbe
# Extra information
plt.title('figure 1') # titre
plt.xlabel('axe x') # axes
plt.ylabel('axe y') # axes
plt.legend() # legend

plt.savefig('figure.png') # sauvegarde la figure dans le repertoire de travail
plt.show() # affiche la figure
```

## Subplots
Pour placer des graphiques les uns à côté des autres, on utilise la fonction subplot(). Les graphiques seront placés comme dans une matrice, avec un nombre de lignes n_lignes et un nombre de colonnes n_colones. On peut indiquer les dimensions de cette matrice en paramètres de la fonction subplot(), à l’aide de la syntaxe suivante :

où courant indique l’indice du graphique actif. Voyons à travers un exemple le fonctionnement :

```python
x = np.arange(-10, 11)
y = -x**2
# Matrice de dimension 3x2 de graphiques

# Ligne 1, colonne 1
plt.subplot(3, 2, 1)
plt.plot(x, y, color = "red")

# Ligne 1, colonne 2
plt.subplot(3, 2, 2)
plt.plot(x, y, color = "orange")

# Ligne 2, colonne 1
plt.subplot(3, 2, 3)
plt.plot(x, y, color = "yellow")

# Ligne 2, colonne 2
plt.subplot(3, 2, 4)
plt.plot(x, y, color = "green")

# Ligne 3, colonne 1
plt.subplot(3, 2, 5)
plt.plot(x, y, color = "blue")

# Ligne 3, colonne 2
plt.subplot(3, 2, 6)
plt.plot(x, y, color = "violet")
```

## Histogrammes
Pour réaliser un histogramme avec pyplot, on utilise la fonction hist() :

```python
x = np.random.randn(1000)
plt.hist(x);
```

On peut préciser avec le paramètre bins soit le nombre de classes, soit leurs bornes :

```python
plt.hist(x, bins=50);
```

Et avec les bornes :

```python
bins = np.arange(-4, 4, .1);
plt.hist(x, bins=bins);
```

L’orientation se change via le paramètre orientation, en indiquant soit 'vertical' (par dédaut), soit 'horizontal'

```python
plt.hist(x, orientation='horizontal');
```

## Boxplots
Pour réaliser une boîte à moustaches (ou boxplot), pyplot offre la fonction boxplot() :

```python
x = np.random.randn(1000)
plt.boxplot(x);
```

En renseigant à False la valeur du paramètre vert, le boxplot est tracé horizontalement :

```python
plt.boxplot(x, vert = False);
```

Réalisez de beaux graphiques avec Seaborn

Seaborn est une librairie qui vient s'ajouter à Matplotlib, remplace certains réglages par défaut et fonctions, et lui ajoute de nouvelles fonctionnalités.
* Ressources : https://seaborn.pydata.org/

```python
import seaborn as sns
```

```python
#Pour voir les relations entre ces caractéristiques, on peut faire des graphiques par paire :
sns.pairplot(data, hue='survived', height=2.5);
```

```python
#Pour avoir une idée sur la correlation entre la variable suvived et les autres variables
plt.figure(figsize=(10,5))
sns.heatmap(data.corr()[['survived']],cmap="RdBu_r",center=0.0, annot=True);
```

```python
# pour voir les variables qui ont des valeurs manquantes
plt.figure(figsize=(20,10))
sns.heatmap(data.isna(), cbar=False)
```

```python

sns.clustermap(data.corr())
```

```python
# Nous pouvons voir la distribution jointe de deux variables :
with sns.axes_style('white'):
    sns.jointplot("fare", "age", data=data, kind='reg')
```
